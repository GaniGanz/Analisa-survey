```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')


# Display the first few rows of the dataframe
df.head()
```

```python
# Menghitung proporsi jawaban dari kolom Q10
q10_proporsi = df['Q10'].value_counts(normalize=True) * 100

# Membuat pie chart
plt.figure(figsize=(8, 6))
plt.pie(q10_proporsi, labels=q10_proporsi.index, autopct='%1.1f%%', startangle=140)
plt.title('Proporsi Jawaban Pertanyaan Q10')
plt.axis('equal')  # Memastikan lingkaran berbentuk lingkaran
plt.show()

# Menampilkan nilai proporsi secara nominal
print("Proporsi Jawaban Pertanyaan Q10:")
print(q10_proporsi)
```

```python
# Menampilkan informasi tentang responden berdasarkan umur
print("Informasi Responden berdasarkan Umur:")
print(df['Umur'].value_counts())
print("\n")

# Menampilkan informasi tentang responden berdasarkan gender
print("Informasi Responden berdasarkan Gender:")
print(df['Gender'].value_counts())
print("\n")

# Menampilkan informasi tentang responden berdasarkan jurusan
print("Informasi Responden berdasarkan Jurusan:")
print(df['Jurusan'].value_counts())
print("\n")


```

```python
df['Q10'] = df['Q10'].apply(lambda x: 1 if x == 'Ya' else 0)
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(15, 8))
for i in range(1, 10):
    plt.subplot(3, 3, i)
    sns.histplot(df[f'Q{i}'], bins=5, kde=True)
    plt.title(f'Distribusi Q{i}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.countplot(x='Umur', data=df)
plt.title('Distribusi Umur')

plt.subplot(1, 3, 2)
sns.countplot(x='Jurusan', data=df)
plt.title('Distribusi Jurusan')

plt.subplot(1, 3, 3)
sns.countplot(x='Gender', data=df)
plt.title('Distribusi Gender')

plt.tight_layout()
plt.show()
```

```python
# Visualisasi distribusi nilai untuk setiap pertanyaan
plt.figure(figsize=(15, 8))
for i in range(1, 10):
    plt.subplot(3, 3, i)
    sns.histplot(df[f'Q{i}'], bins=5, kde=True)
    plt.title(f'Distribusi Q{i}')
plt.tight_layout()
plt.show()

# Visualisasi distribusi nilai untuk setiap pertanyaan berdasarkan Gender
plt.figure(figsize=(15, 12))
for i in range(1, 10):
    plt.subplot(3, 3, i)
    sns.histplot(df, x=f'Q{i}', hue='Gender', bins=5, kde=True, multiple='stack')
    plt.title(f'Distribusi Q{i} berdasarkan Gender')
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 12))
for i in range(1, 10):
    plt.subplot(3, 3, i)
    sns.histplot(df, x=f'Q{i}', hue='Jurusan', bins=5, kde=True, multiple='stack')
    plt.title(f'Distribusi Q{i} berdasarkan Gender')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Menghitung nilai rata-rata dari semua jawaban (Q1 hingga Q9) untuk setiap responden
df['Kesiapan_Kerja'] = df[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9']].mean(axis=1)

# Visualisasi tingkat kesiapan kerja berdasarkan Gender
plt.figure(figsize=(8, 6))
ax = sns.barplot(x='Gender', y='Kesiapan_Kerja', data=df)
plt.title('Tingkat Kesiapan Kerja berdasarkan Gender')
plt.xlabel('Gender')
plt.ylabel('Rata-rata Tingkat Kesiapan Kerja')

# Menambahkan angka di atas bar
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=11, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()

# Visualisasi tingkat kesiapan kerja berdasarkan Jurusan
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Jurusan', y='Kesiapan_Kerja', data=df)
plt.title('Tingkat Kesiapan Kerja berdasarkan Jurusan')
plt.xlabel('Jurusan')
plt.ylabel('Rata-rata Tingkat Kesiapan Kerja')
plt.xticks(rotation=45)

# Menambahkan angka di atas bar
for p in ax.patches:
    ax.annotate(f"{p.get_height():.2f}", (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=11, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()

# Visualisasi tingkat kesiapan kerja berdasarkan Umur
plt.figure(figsize=(10, 6))
sns.stripplot(x='Umur', y='Kesiapan_Kerja', data=df, jitter=True)
plt.title('Tingkat Kesiapan Kerja berdasarkan Umur')
plt.xlabel('Umur')
plt.ylabel('Tingkat Kesiapan Kerja')
plt.show()
```
